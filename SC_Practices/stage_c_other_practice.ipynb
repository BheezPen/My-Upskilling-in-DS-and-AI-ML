```python
def data_quick_view():
    for column in data.columns:
        unique_values = data[column].unique()
        print(f'Unique values in column "{column}": {len(unique_values)}    They are {unique_values}')
        value_counts = data[column].value_counts()
        print(f'Value counts for column "{column}"\n{value_counts}\n\n')
data_quick_view()
```

```python
data['TotalCharges'] = data['TotalCharges'].replace(' ', np.nan, regex=True)
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'])
```

```python
data.info()
```

```python
# Lets start  encoding the column label manually, manually because they are all x parameter except from Church value according to the dataset problem statement

gender_map = {'Male': 1, 'Female': 0}
data['gender'] = data['gender'].replace(gender_map)
```

```python
partner_map = {'Yes': 1, 'No': 0}
data['Partner'] = data['Partner'].replace(partner_map)
```

```python
dependent_map = {'Yes': 1, 'No': 0} #Dependent being 1 means More Chance of continious patronage and should consider keeping.
data['Dependents'] = data['Dependents'].replace(dependent_map)
```

```python
phoneservice_map = {'Yes': 1, 'No': 0} #CLient that signed up for Phone Service
data['PhoneService'] = data['PhoneService'].replace(phoneservice_map)
```

```python
multipleline_map = {'Yes': 1, 'No': 0, 'No phone service' : 0} #No and No Phone Service turned to 0
data['MultipleLines'] = data['MultipleLines'].replace(multipleline_map)
```

```python
internet_map = {'DSL': 1, 'Fiber optic': 1, 'No': 0} #DSL, Fiber optic are positive move
data['InternetService'] = data['InternetService'].replace(internet_map)
```

```python
yes_no_map = {'Yes': 1, 'No': 0, 'No internet service' : 0} #CLient that signed up for Phone Service
data['OnlineSecurity'] = data['OnlineSecurity'].replace(yes_no_map)
data['OnlineBackup'] = data['OnlineBackup'].replace(yes_no_map)
data['DeviceProtection'] = data['DeviceProtection'].replace(yes_no_map)
data['TechSupport'] = data['TechSupport'].replace(yes_no_map)
data['StreamingTV'] = data['StreamingTV'].replace(yes_no_map)
data['StreamingMovies'] = data['StreamingMovies'].replace(yes_no_map)
data['PaperlessBilling'] = data['PaperlessBilling'].replace(yes_no_map)
```

```python
churn_map = {'Yes': 1, 'No': 0}
data['Churn'] = data['Churn'].replace(churn_map)
```

```python
contract_map = {'Month-to-month': 1, 'One year': 2, 'Two year' : 3}
data['Contract'] = data['Contract'].replace(contract_map)
```

```python
payment_method_map = {'Electronic check': 1,'Mailed check': 2, 'Bank transfer (automatic)': 3, 'Credit card (automatic)' : 4}
data['PaymentMethod'] = data['PaymentMethod'].replace(payment_method_map)
```
