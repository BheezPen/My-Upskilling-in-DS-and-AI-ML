```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
raw_data_df = pd.read_csv('ENB2012_data.csv')

raw_data_df
```

```python
raw_data_df.shape
```

```python
raw_data_df.describe()
```

```python
# Will like to rename the column name from x1 to x8 to be more descriptive

# Setting a variable to store the new names in pair with old names all in a dictionary

new_column_names = {'X1':'Relative_Compactness',
                    'X2':'Surface_Area',
                    'X3':'Wall_Area',
                    'X4':'Roof_Area',
                    'X5':'Overall_Height',
                    'X6':'Orientation',
                    'X7':'Glazing_Area',
                    'X8':'Glazing_Area_Distribution',
                    'Y1':'Heating_Load',
                    'Y2':'Cooling_Load',
                   }

new_column_names
```

```python
# Now let me rename the column in the raw_data_df dataframe with the value of the key as in this dictionary to the dataframe dataset

new_data_df = raw_data_df.rename(columns = new_column_names)

new_data_df #Echoed it to confrim it's done
```

```python
# Echoing the dataset describe just top confrim if it has changes successfully
new_data_df.describe
```

```python
simple_linear_reg_df = new_data_df[['Relative_Compactness', 'Cooling_Load']].sample(15, random_state=2)
# Using the sample method means it returns a random sample of item from an axis of object

# So, in this case, it returns 15 random from specified column in the main dataset 
# the random_state value tell how the random datas should be selected and should be fixed and not shuffled.

simple_linear_reg_df
```

```python
#regression plot
# This plot the x axis with the y axis of the 15 axis data expunged from the main dataset 
sns.regplot(x="Relative_Compactness", y="Cooling_Load", data=simple_linear_reg_df)
# THe plotting plots is possible with the seaborn modules imported as sns
```

```python
# After exploring our dataset, lets build a model

#Firstly, we normalise our dataset to a common scale using the min max scaler
from sklearn.preprocessing import MinMaxScaler


scaler = MinMaxScaler()

normalised_df = pd.DataFrame(scaler.fit_transform(new_data_df), columns= new_data_df.columns)

features_df = normalised_df.drop(columns=['Heating_Load', 'Cooling_Load'])

heating_target = normalised_df['Heating_Load']

heating_target
```

```python
# Now, we split our datasets into the training and testing datasets. Recall that we
# had earlier segmented the features and target variables.

from sklearn.model_selection import train_test_split #This give access to divide our data

x_train, x_test, y_train, y_test = train_test_split(features_df, heating_target, test_size=0.3, random_state=1)

x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Instantiating the the LinearRegression from the the LinearRegression Class
from sklearn.linear_model  import LinearRegression

linear_model = LinearRegression()


#fit the model to the training dataset
linear_model.fit(x_train, y_train)
```

```python
# Lets predict with the xtest data
predicted_values = linear_model.predict(x_test)
predicted_values
```

```python
y_test_list = list(y_test)
y_test_list
```

```python
for x in range(1,int(len(predicted_values))):
    print(f'Original Value is {y_test_list[x-1]}, Predicted value is {predicted_values[x-1]}')
```

## Now from the above, it's clear that the prediction is closer and fairly accurate
## Now using the MAE (Mean Absolute Error), let's evaluate the performance of the LinearRegression Model

```python
#Using the MAE evaluation metrics

from sklearn.metrics import mean_absolute_error #Importing the MAE for evaluation
# Now, Lets calculate the MAE value
mae = mean_absolute_error(y_test, predicted_values)

round(mae, 3) #Used the round function to round it up to 3 decimal digit
```

### From the result, it tell that the model is quite good.
### Since low MAE value close to 0 means better model and we got 0.063, we are pretty close to perfect. 

```python
## Using another evaluation metrics call RSS (Residual SUm of Squares)

## Using Residual Sum of Squares (RSS)
import numpy as np  #Importing the numpy for evaluation

#Here, Its simply every predicted value from y_test
## Then look for its square
## Then sum it up 
rss = np.sum(np.square(y_test - predicted_values))

round(rss, 3) #Used the round function to round it up to 3 decimal digit
```

```python
# Root Mean Square Error (RMSE)
from sklearn.metrics import mean_squared_error # Imported the MSE from metrics in sklearn

#The evaluation is done here
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))

round(rmse, 3) #Used the round function to round it up to 3 decimal digit
```

```python
# Using the R-Squared
# NB: In Sklearn (and moststatictical and ml libraries), R-Squared is often referrred to as:
# R-Squared Score
# or R2_score
from sklearn.metrics import r2_score # Import the R_Score 

#r2_score functions does the calculation
r2_score = r2_score(y_test, predicted_values)

round(r2_score, 3) #Used the round function to round it up to 3 decimal digit
```
