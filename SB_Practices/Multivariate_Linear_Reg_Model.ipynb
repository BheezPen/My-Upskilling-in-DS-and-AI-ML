```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
raw_data_df = pd.read_csv('ENB2012_data.csv')

raw_data_df
```

```python
raw_data_df.shape
```

```python
raw_data_df.describe()
```

```python
# Will like to rename the column name from x1 to x8 to be more descriptive

# Setting a variable to store the new names in pair with old names all in a dictionary

new_column_names = {'X1':'Relative_Compactness',
                    'X2':'Surface_Area',
                    'X3':'Wall_Area',
                    'X4':'Roof_Area',
                    'X5':'Overall_Height',
                    'X6':'Orientation',
                    'X7':'Glazing_Area',
                    'X8':'Glazing_Area_Distribution',
                    'Y1':'Heating_Load',
                    'Y2':'Cooling_Load',
                   }

new_column_names
```

```python
# Now let me rename the column in the raw_data_df dataframe with the value of the key as in this dictionary to the dataframe dataset

new_data_df = raw_data_df.rename(columns = new_column_names)

new_data_df #Echoed it to confrim it's done
```

```python
# Echoing the dataset describe just top confrim if it has changes successfully
new_data_df.describe
```

```python
simple_linear_reg_df = new_data_df[['Relative_Compactness', 'Cooling_Load']].sample(15, random_state=2)
# Using the sample method means it returns a random sample of item from an axis of object

# So, in this case, it returns 15 random from specified column in the main dataset 
# the random_state value tell how the random datas should be selected and should be fixed and not shuffled.

simple_linear_reg_df
```

```python
#regression plot
# This plot the x axis with the y axis of the 15 axis data expunged from the main dataset 
sns.regplot(x="Relative_Compactness", y="Cooling_Load", data=simple_linear_reg_df)
# THe plotting plots is possible with the seabon modules imported as sns
```
